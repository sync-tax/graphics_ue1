## A function to create a uniform color image

Let's create a function to create an image filled with a uniform color. <br>
We can use this function to create a background image for our other tasks.

```python
import ipr

# make_image (long version)
def make_image(width: int, height: int, color: list = [0, 0, 0]) -> list:
    """
    Create a new image of the given width and height.
    The image is filled with color.
     Args:
        width (int): width of the image
        height (int): height of the image
        color (list, optional): uniform color for initing the image. Defaults to [0,0,0] (black).

    Returns:
        list: the new image
    """
    
    pixels = []
    for r in range(height):  # rows
        row = [] # create a new list "row"
        for c in range(width):  # columns
            #row.append(color[c % len(color)]) # ensures that rainbow colors restarts the cycle once c exeeds amount of colors in the list
            row.append(color) 
        pixels.append(row)
    return pixels
    
# rainbow colors - might be a shitty idea tho. edit: jup
rainbow_colors = [
    [255, 0, 0],  
    [255, 127, 0], 
    [255, 255, 0], 
    [0, 255, 0],   
    [0, 0, 255],   
    [75, 0, 130],  
    [148, 0, 211]  
]

img = make_image(8, 10)
ipr.imshow(f"empty image {len(img[0])}x{len(img)}", img)
print(img[1])
```

## _Task 02_: Implement a function to create a horizontal gradient image

The function `make_gradient_image` creates an image of a specific width and height. 
The image contains a  horizontal gradient (from left to right). `color1` is on the left side of the image, `color2` is on the right edge of the image. In-between pixels are interpolated linearly between the two colors (e.g., 50% of color1 and 50% of color2 in the center).

The function can create a vertical and horizontal gradient. Think about ways to implement this function in a way that you can reuse as much code as possible.

```python
def make_gradient_image(width:int, height:int, color1:list=[0,0,0], color2:list=[255,255,255]) -> list:
    """Create a new image of the given width and height.
    The image is filled with a gradient between color1 and color2.

    Args:
        width (int):  width of the image
        height (int): height of the image
        color1 (list): gradient color1. Default is black ([0,0,0]).
        color2 (list): gradient color2. Default is white ([255,255,255]).

    Returns:
        list: gradient image
    """
   
    # Todo: implement this function
    img = []
    for v in range(height):  # rows
        row = [] # create a new list "row"
        for u in range(width):  # columns
            ratio = u / (width - 1)  # calc interpolation ratio -> 0 to 1
            pixel_color = [] # declare empty list to store the pixel color values
            for i in range(3): #loop over rgb values 
                pixel_value = int(color1[i] + ratio * (color2[i] - color1[i])) # perfom linear interpolation calc on each channel (interpolated_val=start_val+ratio×(end_val−start_val)
                pixel_color.append(pixel_value) # add to previously initialized list
                #if u == 0:
                    #print(pixel_value)
                    
            #print(pixel_color)
            row.append(pixel_color) #append each pixel color to the row
            #print(row)
        img.append(row) # append each row to img
    return img
    # return make_image(width, height, color1) # I don't really get why i would call the make_img function here (?)

# Test cases for make_gradient_image
img = make_gradient_image(16, 3, [0,0,255], [255,255,0]) 
ipr.imshow(f"gradient image {len(img[0])}x{len(img)}", img) # image should be a gradient from yellow on the left to blue on the right!
```

## _Task 03_: Implement a function to create a checkerboard image

Implement the function `make_checkerboard_image` with the parameters as defined below.
The resulting image shows a checkerboard (the board you know from the game chess) and squares using `color1` and `color2` alternatingly. 
Additionally the size of the squares can be controlled by the parameter `checker_size`. A square size of N means that one square is NxN pixels large.


```python
def make_checkerboard_image(width:int, height:int, color1:list=[0,0,0], color2:list=[255,255,255], checker_size:int=1) -> list:
    """Create a new checkerboard image of the given width and height. 
    The squares of the checkerboard are pixels of colors color1 and color2 and the checkersize determines the size (in pixels) of the squares.

    Args:
        width (int):  width of the image
        height (int): height of the image
        color1 (list, optional): color of even squares (e.g., 0,0). Defaults to [0,0,0].
        color2 (list, optional): color of uneven squares (e.g., 1,0). Defaults to [255,255,255].
        checker_size (int, optional): size of the checkerboard squares. Defaults to 1.

    Returns:
        list: the checkerboard image
    """
    # Todo: implement this function (overwrite the line below)
    img = []
    for v in range(height):  # rows
        row = []
        for u in range(width):  # columns
            checker_index = (u // checker_size) + (v // checker_size)
            #print(square)
            if checker_index % 2 == 0:
                row.append(color1)
            else:
                row.append(color2)
        img.append(row)
    return img
    #return make_image(width, height, color1)

# Test cases for make_checkerboard_image
chess = make_checkerboard_image(16, 8, [0,0,255], [255,255,0]) 
ipr.imshow(f"checkerboard yellow/blue {len(chess[0])}x{len(chess)}", chess) # image is a yellow/blue checkerboard. The left top square is blue. 
chess = make_checkerboard_image(16, 8, checker_size=4)
ipr.imshow(f"checkerboard black/white with large squares {len(chess[0])}x{len(chess)}", chess) # image is a black/white checkerboard, where one square occupies 4 pixels. 
```
